```r
#Starting B. P. will be BP initial
BPI = c(155,142,145,160,149,152,157,159,166,163,158,161)
```

```r
#B.P. After XYZ will be BP final
BPF = c(152,142,144,159,150,153,156,160,165,162,159,160)
```

```r
t.test(BPI, BPF, alternative = c("two.sided"), paired=TRUE)
# p-value greater thatn .05 not significantly different
```
